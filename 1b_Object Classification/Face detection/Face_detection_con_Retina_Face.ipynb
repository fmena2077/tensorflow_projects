```python
import os
import urllib.request
urllib.request.urlretrieve('https://github.com/peteryuX/retinaface-tf2/blob/master/photo/0_Parade_marchingband_1_149.jpg?raw=true', 'marchingband.jpg')
```

```python
os.listdir()
```

```python
from IPython.display import display
from PIL import Image
img_PIL = Image.open(r'marchingband.jpg')
display(img_PIL)
```

```python
from retinaface import RetinaFace
resp = RetinaFace.detect_faces("marchingband.jpg")
```

```python
#Toda la informaciÃ³n que deseamos estÃ¡ presente
resp
```

```python
faces = RetinaFace.extract_faces(img_path = "marchingband.jpg", align = True)
for face in faces:
  plt.imshow(face)
  plt.show()
```
